```
import sys
from PIL import Image
import requests
from io import BytesIO
```

```
response = requests.get("https://www.nicepng.com/png/detail/424-4248926_deluxe-minnie-mouse-cartoon-face-mickey-mouse-clip.png")
#response = requests.get("https://i.pinimg.com/736x/01/86/b5/0186b561bb497897b072ef381509739f.jpg")
img = Image.open(BytesIO(response.content))
w,h = img.size
ar = h/float(w)
nw = 100
nh = int(ar*nw)
im = img.resize((nw,nh))

px = list(im.getdata())
pxl = [px[i:i+im.width] for i in range(0, len(px), im.width)]
avgmat = []
for row in pxl:
  r = []
  for pix in row:
    avg = (pix[0] + pix[1] + pix[2])/3
    r.append(avg)
    
  avgmat.append(r)
asc = "`^\",:;Il!i~+_-?][}{1)(|\\/tfjrxnuvczXYUJCLQ0OZmwqpdb_hao*#MW&8_B@$"
ascmat = []
for row in avgmat:
  ascmatr = []
  for p in row:
    ascmatr.append(asc[int(p/255 * len(asc)) -1])
  ascmat.append(ascmatr)

for row in ascmat:
  for p in row:
        print(p, end = '')
  print("-")
```
